

# Qual estação do ano contém a maior média de locações de bicicletas?

```python
#season: estação do ano (1: inverno, 2: primavera, 3: verão, 4: outono)
# Suponha que você tenha um DataFrame chamado 'df' com uma coluna 'coluna_alvo' e uma coluna 'data'.

# Converta a coluna 'data' para o tipo datetime
df['datetime'] = pd.to_datetime(df['datetime'])

# Estamos criando uma nova coluna chamada "locacoes" - é soma das casuais e registradas
df['locacoes'] = df['casual'] + df['registered']

# Criando uma função para mapear as datas para estações do ano
def season_of_date(date):
    if date.month in [3, 4, 5]:
        return 'Primavera'
    elif date.month in [6, 7, 8]:
        return 'Verão'
    elif date.month in [9, 10, 11]:
        return 'Outono'
    else:
        return 'Inverno'

# Aplicando a função e criando uma nova coluna 'estacao'
df['season'] = df['datetime'].apply(season_of_date)

# Calculando a média de locações para cada estação
media_por_estacao = df.groupby('season')['locacoes'].mean().reset_index()

# Encontrando a estação com a maior média
estacao_maior_media = media_por_estacao.loc[media_por_estacao['locacoes'].idxmax()]

print("Estação com a maior média de locações:", estacao_maior_media['season'])
print("Média de locações:", estacao_maior_media['locacoes'])
```

# Qual estação do ano contém a menor média de locações de bicicletas?

```python
# Suponha que você tenha um DataFrame chamado 'df' com as informações das locações e datas.
# Certifique-se de que a coluna 'data' seja do tipo datetime.
# Suponha também que a coluna de locações seja chamada 'locacoes'.

# Converta a coluna 'data' para o tipo datetime
df['datetime'] = pd.to_datetime(df['datetime'])

# Estamos criando uma nova coluna chamada "locacoes" - é soma das casuais e registradas
df['locacoes'] = df['casual'] + df['registered']

# Criando uma função para mapear as datas para estações do ano
def season_of_date(date):
    if date.month in [3, 4, 5]:
        return 'Primavera'
    elif date.month in [6, 7, 8]:
        return 'Verão'
    elif date.month in [9, 10, 11]:
        return 'Outono'
    else:
        return 'Inverno'

# Aplicando a função e criando uma nova coluna 'estacao'
df['season'] = df['datetime'].apply(season_of_date)

# Calculando a média de locações para cada estação
media_por_estacao = df.groupby('season')['locacoes'].mean().reset_index()

# Encontrando a estação com a menor média
estacao_menor_media = media_por_estacao.loc[media_por_estacao['locacoes'].idxmin()]

print("Estação com a menor média de locações:", estacao_menor_media['season'])
print("Média de locações:", estacao_menor_media['locacoes'])
```

# Às quartas-feiras qual horário com a maior media de locações?

```python
# weekday = 3 (quarta)
```

```python
# Suponha que você tenha um DataFrame chamado df com colunas 'data' e 'locacoes'
# Substitua 'df' pelo nome do seu DataFrame, se necessário

# Certifique-se de que a coluna 'data' seja do tipo datetime
df['datetime'] = pd.to_datetime(df['datetime'])

# Extraia o dia da semana para cada data
df['dia_semana'] = df['datetime'].dt.day_name()

# Filtra apenas as quartas-feiras
quartas_feiras = df[df['dia_semana'] == 'Wednesday']

# Agrupe os dados das quartas-feiras pelo horário e calcule a média de locações para cada horário
media_por_horario_quartas_feiras = quartas_feiras.groupby('hour')['locacoes'].mean()

# Encontre o horário com a maior média de locações nas quartas-feiras
horario_maior_media_quartas_feiras = media_por_horario_quartas_feiras.idxmax()

# Agora você tem o horário com a maior média de locações nas quartas-feiras
print("Horário com a maior média de locações nas quartas-feiras:", horario_maior_media_quartas_feiras)
```
